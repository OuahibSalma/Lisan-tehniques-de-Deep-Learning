```python
import torch
from TTS.api import TTS

# Get device
device = "cuda" if torch.cuda.is_available() else "cpu"

# List available 🐸TTS models
print(TTS().list_models())

# Init TTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)
```

```python
import gradio as gr

# Function to process the initial document
def upload_document(file):
    return "Document uploaded: " + file.name

# Function to process the audio conversation
def process_audio(audio, document):
    # Text to speech list of amplitude values as output
    wav = tts.tts(text="""مخاطر السمنة متعددة وتشمل عدة جوانب صحية واجتماعية. وفقًا للتقرير المقدم، يمكن تلخيص المخاطر كما يلي:

الأمراض المزمنة:

زيادة خطر الإصابة بالسكري من النوع الثاني: يرتبط الوزن الزائد والسمنة بزيادة خطر الإصابة بمرض السكري، خاصة عند الأطفال الذين يولدون لأمهات يعانين من السمنة【​​
أمراض القلب والأوعية الدموية: تتزايد احتمالية الإصابة بأمراض القلب نتيجة للسمنة​​
التأثيرات السلبية على الصحة العامة:

زيادة معدل الوفيات: تتسبب السمنة في ارتفاع معدل الوفيات بسبب الأمراض المزمنة المرتبطة بها.
انخفاض جودة الحياة: يعاني الأفراد الذين يعانون من السمنة من انخفاض في نوعية الحياة بسبب المشاكل الصحية المستمرة​​
التأثيرات على الأجيال القادمة:

تأثير سلبي على الأطفال: السمنة لدى الأمهات تزيد من خطر ولادة أطفال مصابين بالسمنة أو لديهم مشاكل صحية متعلقة بالغذاء​​
التأثيرات النفسية والاجتماعية:

الوصمة الاجتماعية: يعاني الأشخاص الذين يعانون من السمنة من وصمة اجتماعية، مما يؤثر على صحتهم النفسية.
المشاكل النفسية: تشمل هذه المشاكل الاكتئاب، والقلق، وانخفاض الثقة بالنفس.
زيادة الأعباء الصحية والاقتصادية:

ارتفاع تكاليف الرعاية الصحية: تتطلب السمنة والأمراض المرتبطة بها تكاليف علاجية عالية، مما يضيف أعباء اقتصادية على الأفراد والنظم الصحية​​
بشكل عام، تمثل السمنة مشكلة صحية عامة تؤثر على الأفراد والمجتمعات بطرق متعددة، وتستدعي تدابير وقائية وعلاجية للحد من تأثيراتها السلبية.""", speaker_wav="الشخصية_الهزلية.ogg", language="ar")
# Text to speech to a file
    tts.tts_to_file(text="""مخاطر السمنة متعددة وتشمل عدة جوانب صحية واجتماعية. وفقًا للتقرير المقدم، يمكن تلخيص المخاطر كما يلي:

الأمراض المزمنة:

زيادة خطر الإصابة بالسكري من النوع الثاني: يرتبط الوزن الزائد والسمنة بزيادة خطر الإصابة بمرض السكري، خاصة عند الأطفال الذين يولدون لأمهات يعانين من السمنة【​​
أمراض القلب والأوعية الدموية: تتزايد احتمالية الإصابة بأمراض القلب نتيجة للسمنة​​
التأثيرات السلبية على الصحة العامة:

زيادة معدل الوفيات: تتسبب السمنة في ارتفاع معدل الوفيات بسبب الأمراض المزمنة المرتبطة بها.
انخفاض جودة الحياة: يعاني الأفراد الذين يعانون من السمنة من انخفاض في نوعية الحياة بسبب المشاكل الصحية المستمرة​​
التأثيرات على الأجيال القادمة:

تأثير سلبي على الأطفال: السمنة لدى الأمهات تزيد من خطر ولادة أطفال مصابين بالسمنة أو لديهم مشاكل صحية متعلقة بالغذاء​​
التأثيرات النفسية والاجتماعية:

الوصمة الاجتماعية: يعاني الأشخاص الذين يعانون من السمنة من وصمة اجتماعية، مما يؤثر على صحتهم النفسية.
المشاكل النفسية: تشمل هذه المشاكل الاكتئاب، والقلق، وانخفاض الثقة بالنفس.
زيادة الأعباء الصحية والاقتصادية:

ارتفاع تكاليف الرعاية الصحية: تتطلب السمنة والأمراض المرتبطة بها تكاليف علاجية عالية، مما يضيف أعباء اقتصادية على الأفراد والنظم الصحية​​
بشكل عام، تمثل السمنة مشكلة صحية عامة تؤثر على الأفراد والمجتمعات بطرق متعددة، وتستدعي تدابير وقائية وعلاجية للحد من تأثيراتها السلبية.""", speaker_wav="الشخصية_الهزلية.ogg", language="ar", file_path="output.wav")
    return "output.wav"

with gr.Blocks() as demo:
    gr.Markdown("Upload a document and start an audio conversation with the model.")

    # Section to upload the document
    with gr.Row():
        document_input = gr.File(label="Upload Document")
        document_output = gr.Textbox(label="Document Status")

    upload_button = gr.Button("Upload Document")

    # Section for the audio conversation
    with gr.Row():
        audio_input = gr.Audio(sources="microphone", type="filepath", label="Record Audio")
        audio_output = gr.Audio(label="Response Audio")

    submit_button = gr.Button("Submit Audio")

    # Event listeners
    upload_button.click(upload_document, inputs=document_input, outputs=document_output)
    submit_button.click(process_audio, inputs=[audio_input, document_input], outputs=audio_output)

demo.launch()
```
